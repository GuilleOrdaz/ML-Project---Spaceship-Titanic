```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')

df_train = pd.DataFrame(data)
df_train
```

```python
df_train.isnull().sum()
```

```python
df_train=df_train.fillna(df_train.median())
```

```python

df_train['HomePlanet']= df_train['HomePlanet'].fillna(df_train['HomePlanet'].mode().values[0])
df_train['Cabin']= df_train['Cabin'].fillna(df_train['Cabin'].mode().values[0])
df_train['Destination']= df_train['Destination'].fillna(df_train['Destination'].mode().values[0])

df_train.drop('Name',1,inplace=True)
df_train.drop('Cabin',1,inplace=True)
df_train.drop('PassengerId',1,inplace=True)
df_train.drop('Destination',1,inplace=True)
```

```python
origen = pd.get_dummies(df_train)
```

### MODELOS UTILIZADOS

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegressionCV
rdom = RandomForestClassifier(n_estimators=83,criterion = 'gini',min_samples_leaf = 4, min_samples_split=4)
logi = LogisticRegressionCV(random_state=0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
param = {'n_estimators':[100]}
##meter hiperparametros nuevos
gridcv = GridSearchCV(rdom,param_grid = param,cv=5)

gridcv.fit(X_train,y_train)
```

```python
y_hat = gridcv.predict(X_test)
y_hat_train = gridcv.predict(X_train)


from sklearn.metrics import accuracy_score

print('MODELO RANDOM FOREST')
print('Accuracy test' ,accuracy_score(y_test,y_hat))
print('Accuracy train' ,accuracy_score(y_train,y_hat_train))
```

```python
from sklearn.linear_model import LogisticRegressionCV
logi = LogisticRegressionCV(max_iter = 80, random_state=0)
logi.fit(X_train, y_train)

y_hat_logi = logi.predict(X_test)
y_hat_logi_train = logi.predict(X_train)

from sklearn.metrics import accuracy_score

print('MODELO REGRESION LOGÍSTICA')
print('Accuracy test' ,accuracy_score(y_test,y_hat_logi))
print('Accuracy train' ,accuracy_score(y_train,y_hat_logi_train))
```

```python
from xgboost import XGBClassifier
from sklearn import model_selection 

xgbo = XGBClassifier(n_estimators=100)
xgbo.fit(X_train,y_train)
y_hat_boost = xgbo.predict(X_test)
y_hat_boost_tr = xgbo.predict(X_train)


print('MODELO XGBOOST')
print('Accuracy test' ,accuracy_score(y_test,y_hat_boost))
print('Accuracy train' ,accuracy_score(y_train,y_hat_boost_tr))

```

```python
from sklearn.svm import SVC

svc_tr = SVC()
svc_tr.fit(X_train,y_train)
svc_tr.predict(X_test)
y_hat_svc = svc_tr.predict(X_test)
y_hat_train_svc = svc_tr.predict(X_train)

print('MODELO SVC')
print('Accuracy test' ,accuracy_score(y_test,y_hat_svc))
print('Accuracy train' ,accuracy_score(y_train,y_hat_train_svc))
```

```python
parameters = {
    'kernel':('linear', 'rbf', 'poly'), 
    'C':[10, 100], 
    'degree': [2,3],
    'coef0': [-1., 1],
    'gamma': ('scale', 'auto')
    }



clf = GridSearchCV(estimator=svc_tr,
                   param_grid=parameters,
                   n_jobs=-1, # All kernels in computer
                   cv=10,
                   verbose=.5) # Folds in cross validation

clf.fit(X_train, y_train)

print("clf.best_stimator_", clf.best_estimator_)
print("clf.best_params_", clf.best_params_)

# Mean cross-validated score of the best_estimator
print("clf.best_score", clf.best_score_)
```
